```python
class Character:
    strength: int | float
    agility: int | float
    intelligence: int | float
    speed: int | float
    damage: int | float

    
    def __init__(
        self, 
        strength,
        agility,
        intelligence
    ):
        self.strength = strength
        self.agility = agility
        self.intelligence = intelligence
        self.speed = agility * 3.4
        self.damage = strength * 21.3
        
    def trade(self):
        print('trade')

    def wait(self):
        print('wait')

    def think(self):
        print('think')

    def move(self):
        print('move')
        
    def info(self):
        print('Cила:', self.strength)
        print('Ловкость:', self.agility)
        print('Интеллект:', self.intelligence)
        print('Cкорость:', self.speed)
        print('Урон:', self.damage)


# наследование
class Hero(Character):
    name: str = 'красавчик реально'
    angry = False


    def set_name(
        self, 
        name
    ):
        self.name = name

    def info(self):
        print('Имя:', self.name)
        print('Cила:', self.strength)
        print('Ловкость:', self.agility)
        print('Интеллект:', self.intelligence)
        print('Cкорость:', self.speed)
        print('Урон:', self.damage)
        print('Враждебность:', 'злой' if self.angry else 'добрый')
    
    def save(self):
        print(f'{self.name} спас кого-то!!')


# наследование
class Villain(Character):
    name: str = 'плохой парень'
    angry = True


    def set_name(
        self, 
        name
    ):
        self.name = name

    def info(self):
        print('Имя:', self.name)
        print('Cила:', self.strength)
        print('Ловкость:', self.agility)
        print('Интеллект:', self.intelligence)
        print('Cкорость:', self.speed)
        print('Урон:', self.damage)
        print('Враждебность:', 'злой' if self.angry else 'добрый')
    
    def kill(self):
        print(f'{self.name} убил кого-то!!')


# наследование
class Policeman(Hero):
    rank: str = 'рядовой'
    weapon: str = 'дубинка'


    def set_rank(
        self, 
        rank
    ):
        self.rank = rank

    def set_weapon(
        self, 
        weapon
    ):
        self.weapon = weapon
    
    def info(self):
        print('Имя:', self.name)
        print('Должность:', self.rank)
        print('Оружие:', self.weapon)
        print('Cила:', self.strength)
        print('Ловкость:', self.agility)
        print('Интеллект:', self.intelligence)
        print('Cкорость:', self.speed)
        print('Урон:', self.damage)
        print('Враждебность:', 'злой' if self.angry else 'добрый')

    def detain(self):
        print(f'{self.name} задержал преступника с помощью {self.weapon}!!')


# наследование
class Theif(Villain):
    rank: str = 'вор в законе'
    weapon: str = 'заточка'


    def set_rank(
        self, 
        rank
    ):
        self.rank = rank

    def set_weapon(
        self, 
        weapon
    ):
        self.weapon = weapon
    
    def info(self):
        print('Имя:', self.name)
        print('Должность:', self.rank)
        print('Оружие:', self.weapon)
        print('Cила:', self.strength)
        print('Ловкость:', self.agility)
        print('Интеллект:', self.intelligence)
        print('Cкорость:', self.speed)
        print('Урон:', self.damage)
        print('Враждебность:', 'злой' if self.angry else 'добрый')

    def rob(self):
        print(f'{self.name} ограбил человека с помощью {self.weapon}!!')
```

```python
policeman_1 = Policeman(strength=10, agility=20, intelligence=30)
policeman_2 = Policeman(strength=3, agility=17, intelligence=4)

thief_1 = Theif(strength=54, agility=45, intelligence=21)
thief_2 = Theif(strength=23, agility=87, intelligence=3)
```
