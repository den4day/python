```python
class Matrix:
	matrix: list = []

	def __init__(self, matrix):
		self.matrix = matrix

	for value in matrix:
		print(value, ' ')
		print('\n')

	def gram_schmidt(self):
		v1 = self.matrix[0]
		v2 = self.matrix[1]
		v3 = self.matrix[2]
		v4 = self.matrix[3]

		# f1
		f1 = v1
		print('f1 =', f1)
		
		# f2
		f2: list = []
		chislitel1: float = 0
		znamenatel1: float = 0

		for key, value in enumerate(v2):
			chislitel1 += value * f1[key]
			znamenatel1 += f1[key] * f1[key]

		for key, value in enumerate(f1):
			f2.append(v2[key] - chislitel1 / znamenatel1 * f1[key])
			f2[key] = round(f2[key], 3)

		chislitel1 = 0
		znamenatel1 = 0
		print('f2 =', f2)
		
		# f3
		f3: list = []
		
		chislitel2: float = 0
		znamenatel2: float = 0

		for key, value in enumerate(v3):
			chislitel1 += value * f1[key]
			znamenatel1 += f1[key] * f1[key]

			chislitel2 += value * f2[key]
			znamenatel2 += f2[key] * f2[key]

		for key, value in enumerate(f2):
			f3.append(v3[key] - chislitel1 / znamenatel1 * f1[key] - chislitel2 / znamenatel2 * f2[key])
			f3[key] = round(f3[key], 3)

		chislitel1 = 0
		znamenatel1 = 0
		
		chislitel2 = 0
		znamenatel2 = 0

		print('f3 =', f3)

		# f4
		f4: list = []
		
		chislitel3: float = 0
		znamenatel3: float = 0

		for key, value in enumerate(v4):
			chislitel1 += value * f1[key]
			znamenatel1 += f1[key] * f1[key]

			chislitel2 += value * f2[key]
			znamenatel2 += f2[key] * f2[key]

			chislitel3 += value * f3[key]
			znamenatel3 += f3[key] * f3[key]

		for key, value in enumerate(f3):
			f4.append(v4[key] - chislitel1 / znamenatel1 * f1[key] - chislitel2 / znamenatel2 * f2[key] - chislitel3 / znamenatel3 * f3[key])
			f4[key] = round(f4[key], 3)

		chislitel1 = 0
		znamenatel1 = 0
		
		chislitel2 = 0
		znamenatel2 = 0
		
		chislitel3 = 0
		znamenatel3 = 0

		print('f4 =', f4)


		# f1t 
		f1sqrt: float = 0
		f1t: list = []

		# sqrt(f1)
		for value in f1:
			f1sqrt += value * value
		
		f1sqrt = f1sqrt ** 0.5

		print('\nf1sqrt = ', f1sqrt)

		for value in f1:
			f1t.append(round(value / f1sqrt, 3))

		print('f1t =', f1t)


		# f2t 
		f2sqrt: float = 0
		f2t: list = []

		# sqrt(f2)
		for value in f2:
			f2sqrt += value * value
		
		f2sqrt = f2sqrt ** 0.5

		print('\nf2sqrt = ', f2sqrt)

		for value in v2:
			f2t.append(round(value / f2sqrt, 3))

		print('f2t = ', f2t)


		# f3t 
		f3sqrt: float = 0
		f3t: list = []

		# sqrt(f3)
		for value in f3:
			f3sqrt += value * value
		
		f3sqrt = f3sqrt ** 0.5

		print('\nf3sqrt = ', f3sqrt)

		for value in v3:
			f3t.append(round(value / f3sqrt, 3))

		print('f3t =', f3t)


		# f4t 
		f4sqrt: float = 0
		f4t: list = []

		# sqrt(f4)
		for value in f4:
			f4sqrt += value * value
		
		f4sqrt = f4sqrt ** 0.5

		print('\nf4sqrt = ', f4sqrt)

		for value in v4:
			f4t.append(round(value / f4sqrt, 3))

		print('f4t =', f4t)


# Пример использования
matrix = Matrix([[2, -2, 0, 1],
								 [2, 2, 2, 2],
								 [0, -2, 1, 2],
								 [2, 1, -1, 0]])

matrix.gram_schmidt()
```
