```python
class Character:
	def __init__(
			self,
			name: str = 'Имя',
			age: int = 0,
			color: str = 'бежевый',
			isHuman: bool = True
	):
			self.name = name
			self.age = age
			self.color = color
			self.isHumann = isHuman


	def info(self):
			print('Имя:', self.name)
			print('Возраст:', self.age)
			print('Цвет:', self.color)
			print('Человек:', self.isHuman if 'да' else 'нет')
			
	def save(self, name: str = 'человека'):
		print(f'{self.name} cпас {name}!!')


class McQueen(Character):
	name = 'McQueen'
	age = 20
	color = 'красный'
	isHuman = False
	
	maxSpeedMh = 198
	racerNumber = 95
	body = 'купе'


	def info(self):
			print('Имя:', self.name)
			print('Возраст:', self.age)
			print('Цвет:', self.color)
			print('Человек:', self.isHuman if 'да' else 'нет')
			print('Максимальная скорость:', self.maxSpeedMh, 'миль\час')
			print('Гоночный номер:', self.racerNumber)
			print('Тип кузова:', self.body)

	def race(self):
		print(f'{self.name} едет со скоростью {self.maxSpeedMh} миль\час.')

	def enjoyRace(self):
		print(f'{self.name} принимает участие в гонке "Кубок Поршня" c гоночным номером - {self.racerNumber}.')
	
	def wash(self):
		print(f'{self.name} заехал на мойку, чтобы помыть кузов({self.body}).')


class Shrek(Character):
	name = 'Шрек'
	age = 30
	color = 'зелёный'
	isHuman = False


	def info(self):
			print('Имя:', self.name)
			print('Возраст:', self.age)
			print('Цвет:', self.color)
			print('Человек:', self.isHuman if 'да' else 'нет')

	def wash(self):
		print(f'{self.name} принимает грязевые ванны из лечебной глины.')

	def inflate(self):
		print(f'{self.name} надул жабу.')
	
	def scream(self):
		print(f'{self.name} кричит.')


class HarryPotter(Character):
	name = 'Гарри Поттер'
	age = 17
	color = 'бежевый'
	isHuman = True

	haveGlasses = True
	scar = True


	def info(self):
			print('Имя:', self.name)
			print('Возраст:', self.age)
			print('Цвет:', self.color)
			print('Человек:', self.isHuman if 'да' else 'нет')
			print('Носит очки:', self.haveGlasses if 'да' else 'нет')
			print('Имеет шрам:', self.scar if 'да' else 'нет')

	def hex(self):
		print(f'{self.name} применяет заклинание.')

	def quidditch(self):
		print(f'{self.name} играет в Quidditch.')
	
	def parseltongue(self):
		print(f'{self.name} говорит со змеями.')


class FallableMixin():
	def fall(self):
		print('игрушка упала')

class DirtyMixin():
	def dirty(self):
		print('игрушка испачкалась')

class FunkoPopMixin(FallableMixin, DirtyMixin):
	pass


class JumpableMixin():
	def fall(self):
		print('игрушка отскочила')

class TalkableMixin():
	def dirty(self):
		print('игрушка разговаривает')

class PlayableMixin(JumpableMixin, TalkableMixin):
	pass


class PlayableMcQueen(McQueen,  PlayableMixin):
	pass
class FunkoPopMcQueen(McQueen, FunkoPopMixin):
	pass

class PlayableShrek(Shrek,  PlayableMixin):
		pass
class FunkoPopShrek(Shrek, FunkoPopMixin):
		pass

class PlayableHarryPotter(HarryPotter,  PlayableMixin):
		pass
class FunkoPopHarryPotter(HarryPotter, FunkoPopMixin):
		pass
```
